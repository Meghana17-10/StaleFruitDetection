```python
import cv2
import numpy as np
from PIL import Image
import torchvision.transforms as transforms
import matplotlib.pyplot as plt

import os
```

```python
# Load the image
image_path = "/content/monkey-3img.p.jpg"
image = cv2.imread(image_path)

# Convert BGR to RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Convert to PIL image and resize
image_pil = Image.fromarray(image)
image_pil = image_pil.resize((128, 128))

# Display the image
plt.imshow(image_pil)
plt.axis("off")  # Hide axes
plt.show()
```

```python
transform_aug = transforms.Compose([
    transforms.RandomHorizontalFlip(),   # Flip the image horizontally
    transforms.RandomRotation(30),       # Rotate within 30 degrees
    transforms.RandomResizedCrop(128),   # Randomly crop & resize to 128x128
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1)
])
```

```python

fig, axes = plt.subplots(4, 3, figsize=(10, 12))  # Adjust figure size for better visibility

# Generate and display 10 augmented images
for i in range(10):
    row, col = divmod(i, 3)  # Calculate row and column index
    augmented_image = transform_aug(image_pil)  # Apply augmentation
    axes[row, col].imshow(augmented_image)  # Show image
    axes[row, col].axis("off")  # Hide axes

# Hide any unused subplot (last empty one)
axes[3, 2].axis("off")

plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

```python
image_augmented = transform_aug(image_pil)
image_aug_np = np.array(image_augmented) / 255.0
```

```python
# Display original preprocessed and augmented image
fig, axes = plt.subplots(1, 1, figsize=(10, 5))

# Show preprocessed image
# Assuming 'image_pil' is the preprocessed image
axes.imshow(np.array(image_pil) / 255.0)  # Normalize for display
axes.axis("off")
axes.set_title("Preprocessed Image")


plt.show()
```
